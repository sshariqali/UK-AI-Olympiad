## **1. Introduction to Pandas**

### What is Pandas?

Pandas is an open-source Python library that provides high-performance, easy-to-use data structures and data analysis tools. It's built on top of NumPy and is essential for:
- Data cleaning and preparation
- Data analysis and exploration
- Time series analysis
- Statistical analysis

### Core Data Structures

Pandas has two primary data structures:
- **Series**: 1-dimensional labeled array (like a column in Excel)
- **DataFrame**: 2-dimensional labeled data structure (like a spreadsheet or SQL table)

### Importing Pandas

```python
# Standard convention for importing Pandas
import pandas as pd
import numpy as np

# Check version
print(f"Pandas version: {pd.__version__}")
```

## **2. Creating Pandas Objects**

### Creating a Series

```python
# Creating a Series from a list
series_from_list = pd.Series([10, 20, 30, 40, 50])
print("Series from list:")
print(series_from_list)
print(f"\nType: {type(series_from_list)}")
```

```python
# Creating a Series from a dictionary
series_from_dict = pd.Series({
    'Apple': 5,
    'Banana': 3,
    'Orange': 7,
    'Mango': 2
})
print("Series from dictionary:")
print(series_from_dict)
```

### Creating a DataFrame

```python
# Creating a DataFrame from a dictionary of lists
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'City': ['New York', 'Paris', 'London', 'Tokyo', 'Berlin'],
    'Salary': [70000, 80000, 75000, 90000, 85000]
}

df = pd.DataFrame(data_dict)
print("DataFrame from dictionary of lists:")
print(df)
```

```python
# Creating a DataFrame from a list of dictionaries
list_of_dicts = [
    {'Product': 'Laptop', 'Price': 1200, 'Stock': 15},
    {'Product': 'Mouse', 'Price': 25, 'Stock': 50},
    {'Product': 'Keyboard', 'Price': 75, 'Stock': 30},
    {'Product': 'Monitor', 'Price': 300, 'Stock': 20}
]

df_products = pd.DataFrame(list_of_dicts)
print("DataFrame from list of dictionaries:")
print(df_products)
```

### Loading Data from a File

```python
# Loading data from CSV file
df_employees = pd.read_csv('employees.csv')
print("DataFrame loaded from CSV:")
print(df_employees)
```

## **3. Basic DataFrame Inspection**

### Viewing Data

```python
# View first few rows (default is 5)
print("First 3 rows using .head():")
print(df_employees.head(3))
print("\n" + "="*50 + "\n")

# View last few rows
print("Last 2 rows using .tail():")
print(df_employees.tail(2))
```

## **4. Indexing and Selection**

### Selecting Columns

```python
# Select a single column (returns a Series)
print("Single column selection (Name):")
print(df_employees['Name'])
print(f"\nType: {type(df_employees['Name'])}")
```

```python
# Select multiple columns (returns a DataFrame)
print("Multiple column selection:")
print(df_employees[['Name', 'Age', 'Salary']])
print(f"\nType: {type(df_employees[['Name', 'Age', 'Salary']])}")
```

### .iloc - Integer-position Based Selection

Use `.iloc` for selecting by integer position (like array indexing).

```python
# Select a single row by position
print("First row using .iloc[0]:")
print(df_employees.iloc[0])
print("\n" + "="*50 + "\n")

# Select multiple rows
print("Rows 1-3 using .iloc[1:4]:")
print(df_employees.iloc[1:4])
print("\n" + "="*50 + "\n")

# Select specific rows and columns
print("Rows 0-2, Columns 0-2 using .iloc[0:3, 0:3]:")
print(df_employees.iloc[0:3, 0:3])
```

### .loc - Label-based Selection

Use `.loc` for selecting by labels (index and column names).

```python
# Select row by label (index)
print("Row with index 2 using .loc[2]:")
print(df_employees.loc[2])
print("\n" + "="*50 + "\n")

# Select specific rows and columns by label
print("Rows 0-2, specific columns using .loc:")
print(df_employees.loc[0:2, ['Name', 'Department', 'Salary']])
print("\n" + "="*50 + "\n")

# Select all rows, specific columns
print("All rows, Name and Age columns:")
print(df_employees.loc[:, ['Name', 'Age']])
```

## **5. Filtering with Logical Conditions**

### Boolean Masks

```python
# Create a boolean mask
age_mask = df_employees['Age'] > 30
print("Boolean mask for Age > 30:")
print(age_mask)
print(f"\nType: {type(age_mask)}")
```

### Filtering with Single Conditions

```python
# Filter employees older than 30
print("Employees older than 30:")
print(df_employees[df_employees['Age'] > 30])
print("\n" + "="*50 + "\n")

# Filter by department
print("Employees in Engineering:")
print(df_employees[df_employees['Department'] == 'Engineering'])
```

### Filtering with Multiple Conditions

Use `&` for AND and `|` for OR. **Important:** Use parentheses around each condition!

```python
# AND condition: Age > 30 AND Salary > 90000
print("Employees older than 30 AND earning more than 90000:")
print(df_employees[(df_employees['Age'] > 30) & (df_employees['Salary'] > 90000)])
print("\n" + "="*50 + "\n")

# OR condition: Department is Engineering OR Marketing
print("Employees in Engineering OR Marketing:")
print(df_employees[(df_employees['Department'] == 'Engineering') | (df_employees['Department'] == 'Marketing')])
```

### Filtering with .isin()

Use `.isin()` to filter based on a list of values.

```python
# Filter employees in specific departments
departments_of_interest = ['Engineering', 'HR', 'Management']
print(f"Employees in {departments_of_interest}:")
print(df_employees[df_employees['Department'].isin(departments_of_interest)])
```

## **6. Assigning and Modifying Data**

### Creating New Columns

```python
# Create a copy to avoid modifying the original
df_modified = df_employees.copy()

# Add a column with a scalar value
df_modified['Country'] = 'USA'

# Add a column based on calculations from other columns
df_modified['Annual_Bonus'] = df_modified['Salary'] * 0.10

# Add a column based on conditions
df_modified['Seniority'] = df_modified['Age'].apply(lambda x: 'Senior' if x >= 35 else 'Junior')

print("DataFrame with new columns:")
print(df_modified)
```

### Modifying Data Using .loc

```python
# Modify a specific cell
df_modified.loc[0, 'Salary'] = 100000
print("After modifying first employee's salary:")
print(df_modified.head(3))
print("\n" + "="*50 + "\n")

# Modify multiple rows based on condition
df_modified.loc[df_modified['Department'] == 'Marketing', 'Annual_Bonus'] = 15000
print("After updating Marketing bonuses:")
print(df_modified[df_modified['Department'] == 'Marketing'])
```

### âš ï¸ Warning: Chained Indexing

Chained indexing can lead to unexpected behavior. Always use `.loc` or `.iloc` for assignments.

```python
# BAD: Chained indexing (may not work as expected)
# df_modified[df_modified['Age'] > 30]['Salary'] = 100000  # Don't do this!

# GOOD: Using .loc
df_modified.loc[df_modified['Age'] > 30, 'Salary'] = df_modified.loc[df_modified['Age'] > 30, 'Salary'] + 5000
print("Correctly modified using .loc:")
print(df_modified[['Name', 'Age', 'Salary']])
```

## **7. Cleaning & Aggregation**

### Handling Missing Values

First, let's create a DataFrame with missing values.

```python
# Create a DataFrame with missing values
data_with_nan = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, np.nan, 35, 28, np.nan],
    'City': ['New York', 'Paris', np.nan, 'Tokyo', 'Berlin'],
    'Salary': [70000, 80000, 75000, np.nan, 85000]
}

df_nan = pd.DataFrame(data_with_nan)
print("DataFrame with missing values:")
print(df_nan)
```

```python
# Check for missing values
print("Missing values count:")
print(df_nan.isnull().sum())
print("\n" + "="*50 + "\n")

# Drop rows with any missing values
print("After dropping rows with missing values:")
print(df_nan.dropna())
print("\n" + "="*50 + "\n")

# Fill missing values with a specific value
print("After filling missing values:")
df_filled = df_nan.fillna({'Age': df_nan['Age'].mean(), 'City': 'Unknown', 'Salary': 0})
print(df_filled)
```

### Basic Aggregation with .groupby()

GroupBy allows you to split data into groups and apply functions to each group.

```python
# Group by Department and calculate mean salary
print("Average salary by department:")
print(df_employees.groupby('Department')['Salary'].mean())
print("\n" + "="*50 + "\n")

# Multiple aggregations
print("Statistics by department:")
print(df_employees.groupby('Department')['Salary'].agg(['mean', 'min', 'max', 'count']))
print("\n" + "="*50 + "\n")

# Group by and aggregate multiple columns
print("Age and Salary statistics by department:")
print(df_employees.groupby('Department')[['Age', 'Salary']].mean())
```

## **8. Advanced Topics (Optional Preview)**

### Combining DataFrames

```python
# Create sample DataFrames for merging
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Department': ['IT', 'HR', 'IT', 'Sales']
})

df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 5],
    'Salary': [70000, 65000, 75000, 68000],
    'Years': [3, 5, 2, 4]
})

# Merge DataFrames (like SQL JOIN)
print("Merged DataFrame (inner join):")
merged_df = pd.merge(df1, df2, on='EmployeeID', how='inner')
print(merged_df)
print("\n" + "="*50 + "\n")

# Outer join to keep all records
print("Merged DataFrame (outer join):")
merged_outer = pd.merge(df1, df2, on='EmployeeID', how='outer')
print(merged_outer)
```

```python
# Concatenate DataFrames vertically
df_a = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df_b = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

print("Concatenated DataFrames:")
concatenated = pd.concat([df_a, df_b], ignore_index=True)
print(concatenated)
```

### Applying Custom Functions

Use `.apply()` and `.map()` to apply custom transformations.

```python
# Apply a function to a column
def salary_category(salary):
    if salary < 70000:
        return 'Low'
    elif salary < 90000:
        return 'Medium'
    else:
        return 'High'

df_employees['Salary_Category'] = df_employees['Salary'].apply(salary_category)
print("DataFrame with salary categories:")
print(df_employees[['Name', 'Salary', 'Salary_Category']])
print("\n" + "="*50 + "\n")

# Using lambda functions
df_employees['Name_Length'] = df_employees['Name'].apply(lambda x: len(x))
print("DataFrame with name lengths:")
print(df_employees[['Name', 'Name_Length']])
```

```python
# Using .map() with a dictionary
department_mapping = {
    'Engineering': 'Tech',
    'Marketing': 'Business',
    'HR': 'Support',
    'Management': 'Leadership'
}

df_employees['Department_Category'] = df_employees['Department'].map(department_mapping)
print("DataFrame with department categories:")
print(df_employees[['Name', 'Department', 'Department_Category']])
```

### Reshaping Data with .pivot_table()

Pivot tables help reorganize and summarize data.

```python
# Create sample sales data
sales_data = {
    'Date': ['2024-01', '2024-01', '2024-02', '2024-02', '2024-01', '2024-02'],
    'Product': ['Laptop', 'Mouse', 'Laptop', 'Mouse', 'Keyboard', 'Keyboard'],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South'],
    'Sales': [1200, 25, 1500, 30, 75, 80]
}

df_sales = pd.DataFrame(sales_data)
print("Sales data:")
print(df_sales)
print("\n" + "="*50 + "\n")

# Create a pivot table
pivot = df_sales.pivot_table(
    values='Sales',
    index='Product',
    columns='Region',
    aggfunc='sum',
    fill_value=0
)
print("Pivot table - Sales by Product and Region:")
print(pivot)
```

### Working with Text Data (.str accessor)

The `.str` accessor provides string methods for Series.

```python
# String operations
print("Names in uppercase:")
print(df_employees['Name'].str.upper())
print("\n" + "="*50 + "\n")

# String contains
print("Employees with 'son' in their name:")
print(df_employees[df_employees['Name'].str.contains('son', case=False)])
print("\n" + "="*50 + "\n")

# String split
print("First names:")
print(df_employees['Name'].str.split().str[0])
print("\n" + "="*50 + "\n")

# String replace
print("Names with 'John' replaced by 'Jonathan':")
print(df_employees['Name'].str.replace('John', 'Jonathan'))
```

### Working with Datetime

Pandas has powerful datetime functionality for time series analysis.

```python
# Convert string to datetime
df_employees['JoinDate'] = pd.to_datetime(df_employees['JoinDate'])
print("DataFrame with datetime:")
print(df_employees[['Name', 'JoinDate']])
print(f"\nData type of JoinDate: {df_employees['JoinDate'].dtype}")
print("\n" + "="*50 + "\n")

# Extract date components
df_employees['JoinYear'] = df_employees['JoinDate'].dt.year
df_employees['JoinMonth'] = df_employees['JoinDate'].dt.month
df_employees['JoinDayOfWeek'] = df_employees['JoinDate'].dt.day_name()

print("Date components:")
print(df_employees[['Name', 'JoinDate', 'JoinYear', 'JoinMonth', 'JoinDayOfWeek']])
print("\n" + "="*50 + "\n")

# Calculate time differences
df_employees['DaysEmployed'] = (pd.Timestamp.now() - df_employees['JoinDate']).dt.days
print("Days employed:")
print(df_employees[['Name', 'JoinDate', 'DaysEmployed']].sort_values('DaysEmployed', ascending=False))
```
